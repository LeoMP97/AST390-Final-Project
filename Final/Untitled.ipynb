```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import matplotlib as mpl
import seaborn as sns
```

```python
data = pd.read_csv('data_new.csv')  
# print(data)
xpos = data['x-position']
ypos = data['y-position']
ef_r = data['real part of ef']
ef_i = data['imaginary part of ef']
g = data['reduced shear']

fig = plt.figure()
ax = fig.add_subplot(111)
ax.scatter(g,xpos)
ax.scatter(g,ypos)
ax.set_xlabel('Shear')
ax.set_ylabel('position')
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(g, xpos, ypos)

ax.set_xlabel('Reduced Shear')
ax.set_ylabel('x-position')
ax.set_zlabel('y-position')

plt.show()
```
